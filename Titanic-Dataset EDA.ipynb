```python
# ==================================================
# TITANIC DATASET EDA (Project-1)
# ==================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------------------------------
# 1. Load Dataset
# --------------------------------------------------
df = pd.read_csv("Titanic-Dataset.csv")

print("First 5 rows:")
print(df.head())

print("\nDataset Information:")
print(df.info())

# --------------------------------------------------
# 2. Handle Missing Values
# --------------------------------------------------
print("\nMissing Values:\n")
print(df.isnull().sum())

# Fill missing Age with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# --------------------------------------------------
# 3. Explore survival counts
# --------------------------------------------------
plt.figure(figsize=(6,4))
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()

# --------------------------------------------------
# 4. Survival based on Gender
# --------------------------------------------------
plt.figure(figsize=(6,4))
df.groupby('Sex')['Survived'].value_counts().unstack().plot(kind='bar')
plt.title("Survival Based on Gender")
plt.show()

# --------------------------------------------------
# 5. Survival based on Passenger Class
# --------------------------------------------------
plt.figure(figsize=(6,4))
df.groupby('Pclass')['Survived'].value_counts().unstack().plot(kind='bar')
plt.title("Survival Based on Passenger Class")
plt.show()

# --------------------------------------------------
# 6. Age Distribution
# --------------------------------------------------
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# --------------------------------------------------
# 7. Fare Distribution
# --------------------------------------------------
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()

# --------------------------------------------------
# 8. Conclusion
# --------------------------------------------------
print("\nConclusion:")
print("1. More passengers did not survive compared to those who survived.")
print("2. Female passengers had higher survival rate.")
print("3. First class passengers survived more than third class passengers.")
print("4. Most passengers were between age 20 and 40.")
print("5. Fare distribution is highly skewed.")
```
