```python
import pandas as pd
import logging

# -----------------------------
# Setup logging
# -----------------------------
logging.basicConfig(
    filename='cleaning_log.txt',
    level=logging.INFO,
    format='%(asctime)s - %(message)s'
)

# -----------------------------
# 1. Load Dataset
# -----------------------------
file_path = "attendance_2023_2024.csv"

try:
    df = pd.read_csv(file_path)
    logging.info("Dataset loaded successfully")
    print("Dataset Loaded Successfully!\n")
except FileNotFoundError:
    logging.error(f"File not found: {file_path}")
    print(f"Error: File not found: {file_path}")
    exit()

# -----------------------------
# 2. Show first rows
# -----------------------------
print("First 5 Rows:")
print(df.head())

# -----------------------------
# 3. Standardize column names
# -----------------------------
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
logging.info("Column names standardized")

# -----------------------------
# 4. Dataset info
# -----------------------------
print("\nDataset Info:")
print(df.info())
print("\nDataset Shape:", df.shape)

# -----------------------------
# 5. Detect missing values
# -----------------------------
missing = df.isnull().sum()
print("\nMissing Values Before Cleaning:")
print(missing)
logging.info("Checked missing values")

# -----------------------------
# 6. Handle missing values
# -----------------------------
for column in df.columns:
    missing_before = df[column].isnull().sum()
    if missing_before > 0:
        if df[column].dtype in ["int64", "float64"]:
            df[column] = df[column].fillna(df[column].median())  # safer than mean
            logging.info(f"Filled {missing_before} missing numeric values in '{column}' with median")
        else:
            df[column] = df[column].fillna(df[column].mode()[0])
            logging.info(f"Filled {missing_before} missing categorical values in '{column}' with mode")

# -----------------------------
# 7. Remove duplicates
# -----------------------------
duplicates = df.duplicated().sum()
df = df.drop_duplicates()
logging.info(f"Removed {duplicates} duplicate rows")
print(f"\nRemoved {duplicates} duplicate rows")

# -----------------------------
# 8. Attempt numeric type conversion
# -----------------------------
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')
logging.info("Attempted numeric type conversion")

# -----------------------------
# 9. Save cleaned dataset
# -----------------------------
clean_file = "cleaned_attendance_data.csv"
df.to_csv(clean_file, index=False)
logging.info(f"Saved cleaned dataset as '{clean_file}'")
print("\nCleaned dataset saved as:", clean_file)

# -----------------------------
# 10. Completed
# -----------------------------
print("\nData Cleaning Project Completed Successfully!")
```
