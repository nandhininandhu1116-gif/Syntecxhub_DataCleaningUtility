```python
# ==============================
# DATA SCIENCE WEEK 2 PROJECT
# Time Series & Category Charts
# ==============================

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# For displaying graphs in Jupyter Notebook
%matplotlib inline

# ------------------------------
# Create Sample Sales Dataset
# ------------------------------
data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug"],
    "Sales": [2000,2500,3000,2800,3200,3500,3700,4000],
    "Category": ["Electronics","Clothing","Electronics","Furniture",
                 "Clothing","Electronics","Furniture","Electronics"]
}

df = pd.DataFrame(data)

print("Sales Dataset")
print(df)

# ------------------------------
# 1. LINE CHART (Sales Over Time)
# ------------------------------
plt.figure(figsize=(8,5))
plt.plot(df["Month"], df["Sales"], marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)

plt.savefig("line_chart.png")
plt.show()

# ------------------------------
# 2. BAR CHART (Category Comparison)
# ------------------------------
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.savefig("bar_chart.png")
plt.show()

# ------------------------------
# 3. PIE CHART (Category Share)
# ------------------------------
plt.figure(figsize=(6,6))
category_sales.plot(kind="pie", autopct='%1.1f%%')

plt.title("Category Share in Total Sales")
plt.ylabel("")

plt.savefig("pie_chart.png")
plt.show()

# ------------------------------
# SUMMARY
# ------------------------------
print("\nProject Summary:")
print("1. Sales trend shows growth from January to August.")
print("2. Electronics category contributes the highest sales.")
print("3. Bar and pie charts help compare category performance.")
print("4. Line chart clearly shows the sales growth over time.")
```
